```python
def toNumeric(y):
    if not isinstance(y, str):  # If already a number, return it
        return 0.0
    
    y = y.lower().strip()  # Normalize the string
    if "varies" in y:
        return 0.0  # Handle "Varies with device"
    
    # Extract numeric part (digits and dots only)
    num_part = ''.join(c if c.isdigit() or c == '.' else '' for c in y)
    
    try:
        return float(num_part)  # Convert to float
    except ValueError:
        return 0.0  # Default to 0.0 if conversion fails
```
