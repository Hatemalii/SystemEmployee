```python
#Write your code here
x = 0
y=0
z=0
for index in range(len(df)):
    price = df['price'].iloc[index]# Access the price from the row
    if price > x:
        z = y
        y = x
        x = price
    elif price > y:
        z = y
        y = price
    elif price > z:
        z = price
avg = (x+y+z)/3
print(avg)
```
